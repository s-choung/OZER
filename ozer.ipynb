```python
import math
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (24,10)
ax = plt.gca()
#ax.set_xlim([0, 18])
#ax.set_ylim([-0.1, 10])

class MyList(list):
    def __init__(self, *args):
        super(MyList, self).__init__(args)

    def __sub__(self, other):
        return self.__class__(*[item for item in self if item not in other])

subscript = str.maketrans("0123456789-_", "₀₁₂₃₄₅₆₇₈₉→→")
global label_exist1
global label_exist2
label_exist1 = 1
label_exist2 = 1
plt.plot([0,1],[0,0],color = 'black', linewidth=3)
plt.text(0.2,0.1,'H2O'.translate(subscript), fontsize=16)
itms = ['OH', 'O', 'OOH', 'O2'.translate(subscript), 'OH+O2'.translate(subscript), '2OH', 'OH+O', '2O', 'OH+OOH' , 'O+OOH', 'OOHO', 'O+O2'.translate(subscript), 'O3'.translate(subscript), 'O3+H'.translate(subscript), 'O3+V'.translate(subscript), 'O2+V'.translate(subscript), 'H', 'O2(g)'.translate(subscript), 'O3(g)'.translate(subscript)]
ne= [1, 2, 3, 4, 5, 2, 3, 4, 4, 5, 5, 6, 6, 5, 4, 2, 3, 4, 6] # 이동한 전자수.
s = [1.67, 3.70, 4.32, 4.64, 6.23, 3.22, 5.61, 8.00, 5.98, 8.32, 8.15, 8.62, 7.77, 5.11, 6.36, 4.44, 5.86, 4.92, 9.06]
ss = [0.35, 3.04, 3.39, 4.40, 5.28, 1.92,	4.31,	6.80,	4.59,	7.38,	6.72,	7.73, 6.51, 5.23, 6.46, 4.23, 6.73, 4.92, 9.06 ]
nss = [1.65, 3.54, 4.41, 4.50, 6.06, 3.17, 5.19, 7.54, 5.93, 7.94, 8.01, 7.64, 7.73, 5.23, 6.82, 3.96, 4.26, 4.92, 9.06 ]



def volt(list1,list2):
  difference = []
  zip_object = zip(list1, list2)
  for list1_i, list2_i in zip_object:
      difference.append(list1_i-list2_i)
  return difference

def liner(value,position,name,mech):  #G, nth e+H transfer, intermediate,aem(a) or lom(l)
  global label_exist1
  global label_exist2
  if(mech == 'a' and label_exist1 ==1):
    x=[position,position+1]
    l=[value,value]
    plt.plot(x,l,color = 'black', linewidth=3)
    plt.text(position+0.5,value+0.05,name, fontsize=16,ha='center')
  else:
    x=[position,position+1]
    l=[value,value]
    plt.plot(x,l,color = 'red', linewidth=3)
    plt.text(position+0.5,value+0.05,name, fontsize=16,color = 'red',ha='center')
  return


def connector(initial,final,position,mech):#G1 ,G2, nth e+H transfer
  span=0.1
  c_fill1=[initial+span,final+span]
  c_fill2=[initial-span,final-span]
  c=[initial,final]
  x=[position,position+1]
  if(mech == 'a'):
    plt.plot(x,c,'--',color = 'black')
  elif(mech == 'l'):
    plt.plot(x,c,'--',color = 'red')
  elif(mech =='chem_a'):
    plt.plot(x,c,'--',color ='black')
    plt.fill_between(x, c_fill1 ,c_fill2, color = 'yellow', alpha=0.5)
  elif(mech =='chem_l'):
    plt.plot(x,c,'--',color ='red')
    plt.fill_between(x, c_fill1 ,c_fill2, color = 'yellow', alpha=0.5)
  return

def lines(catalyst,v):#surface, applied potential
  voltage = [element * v for element in ne]
  a=catalyst
  c=volt(a,voltage)
  liner(c[14],10,itms[14],'l')
  liner(c[15],6,itms[15],'l')
  liner(c[16],8,itms[16],'l')
#  liner(c[16],14,itms[16],'l')
  liner(c[13],12,itms[13],'l')

  for i in range (len(s)):
    if(i==0):#oh
      liner(c[i],2,itms[i],'a') 
    elif(i == 1 or i == 5):#o 2OH 
      liner(c[i],4,itms[i],'a')  
    elif(i == 2  or i == 6):#ooh o+oh O2+V
      liner(c[i],6,itms[i],'a')  
    elif(i == 3 or i == 8 or i == 7 ):#o2 2O OOH+OH V+O3 H
      liner(c[i],8,itms[i],'a')  
    elif(i == 4 or i == 17  or i == 9 ):#o2(g) O2+OH OOH+O
      liner(c[i],10,itms[i],'a')  
    elif(i == 10 or i == 11 ):# O2+OH OOHO
      liner(c[i],12,itms[i],'a')  
    elif(i == 12 ):
      liner(c[i],14,itms[i],'a')  
    elif( i == 18 ):#o3(g)
      liner(c[i],16,itms[i],'a')  
  return

def connection(catalyst,v):#surface, applied potential
  voltage = [element * v for element in ne]
  a=catalyst
  c=volt(a,voltage)
  connector(0,c[0],1,'a') 

  connector(c[0],c[1],3,'a') 
  connector(c[0],c[5],3,'a') 

  connector(c[1],c[2],5,'a') 
  connector(c[1],c[6],5,'a') 
  connector(c[5],c[6],5,'a') 
  connector(c[1],c[15],5,'chem_l') 

  connector(c[15],c[16],7,'l') 
  connector(c[2],c[3],7,'a') 
  connector(c[6],c[7],7,'a') 
  connector(c[6],c[8],7,'a') 

  connector(c[16],c[17],9,'l') 
  connector(c[3],c[14],9,'chem_l') 
  connector(c[3],c[17],9,'chem_a') 
  connector(c[3],c[4],9,'a') 
  connector(c[8],c[9],9,'a') 
  connector(c[8],c[4],9,'a') 
  connector(c[7],c[9],9,'a') 
  connector(c[7],c[9],9,'a') 

  connector(c[4],c[10],11,'chem_a') 
  connector(c[4],c[11],11,'a') 
  connector(c[9],c[10],11,'chem_a') 
  connector(c[9],c[11],11,'a') 
  connector(c[14],c[13],11,'l') 

  connector(c[13],c[12],13,'l') 
  connector(c[10],c[12],13,'a') 
  connector(c[11],c[12],13,'chem_a') 

  connector(c[12],c[18],15,'chem_a') 

  return

def plotall(cat,vo):
  lines(cat,vo)
  connection(cat,vo)
  return

ax.tick_params(width=3)
plt.margins(x=0.01)
plt.margins(y=0.01)
plt.xlabel('Reaction coordinate', fontsize=16)
plt.ylabel('ΔG (eV)', fontsize=16)
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) # labels along the bottom edge are off
plt.tick_params(    axis='y', labelsize=16       ) # labels along the bottom edge are off
'''
plotall(s,0)
plt.savefig("s.jpg", dpi=150)
plotall(ss,0)
plt.savefig("ss.jpg", dpi=150)'''
plotall(nss,0)
plt.savefig("nss.jpg", dpi=150)
```
